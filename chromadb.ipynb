```python
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_groq import ChatGroq
from langchain_community.document_loaders import DirectoryLoader, TextLoader
```

```python
loader = DirectoryLoader("C:/Users/maaz7/generative ai chromadb/new_articles", 
                         glob = "./*.txt", loader_cls=TextLoader, loader_kwargs={
        "encoding": "utf-8",
        "autodetect_encoding": True
    } )
```

```python
document = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 20)
```

```python
text_chunks = text_splitter.split_documents(document)
```

```python
text_chunks[0].page_content
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embedding = HuggingFaceEmbeddings(
    model_name="BAAI/bge-small-en-v1.5"
)
```

```python
!pip install sentence-transformers
```

```python
persist_directory='db'
```

```python
vectordb = Chroma.from_documents(
    documents=text_chunks,
    embedding=embedding,
    persist_directory=persist_directory
)
```

```python
vectordb=None
```

```python
vectordb = Chroma(persist_directory=persist_directory, embedding_function=embedding)
```

```python
retriever = vectordb.as_retriever()
```

```python
docs = retriever.invoke("How much money did microsoft raise")
```

```python
import langchain
print(langchain.__version__)
```

```python
from langchain_classic.chains import RetrievalQA
from langchain_groq import ChatGroq
from dotenv import load_dotenv
```

```python
load_dotenv()
groq_api_key=os.getenv("GROQ_KEY")
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=groq_api_key
              )
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
def process_llm_response(llm_response):
    print(llm_response['result'])
    print('\n\nSources:')
    for source in llm_response["source_documents"]:
        print(source.metadata['source'])
```

```python
prompt="How much money did microsoft raise?"
```

```python
llm_response = qa_chain.invoke(prompt)
```

```python
llm_response
```

```python
process_llm_response(llm_response)
```
